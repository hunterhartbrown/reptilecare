```python
import pandas as pd
import os

# Paths (customize if needed)
excel_path = r'C:\Users\Hunter\Desktop\Excel\Website\reptilecare\data\Availability.xlsx'
json_path = r'C:\Users\Hunter\Desktop\Excel\Website\reptilecare\data\availability.json'

# Load the Excel file
df = pd.read_excel(excel_path, engine='openpyxl')

# Optional: Clean or select columns if needed, for example:
# df = df[['Date', 'Animal Name', 'MorphMarket # of Listings', 'US & Canada MM Availability']]

# Convert the DataFrame to JSON
json_data = df.to_json(orient='records')

# Make sure the output directory exists
os.makedirs(os.path.dirname(json_path), exist_ok=True)

# Save the JSON data to a file
with open(json_path, 'w', encoding='utf-8') as json_file:
    json_file.write(json_data)

print("Availability JSON data successfully saved.")
```
